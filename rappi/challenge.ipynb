## Understanding Numeric Variables - Active Users

Active users are people has assigned credit cardand has any transaction

```python
active = """
SELECT
ID,
usage_amount,
AMOUNT,
ROUND(usage_amount/AMOUNT,2) as credit_capacity,
(CASE WHEN ROUND(usage_amount/AMOUNT,2)>1 THEN 1 ELSE 0 END) as band_ratio,
nb_of_transactions,
amount_transaction_mean
FROM
(
SELECT
ID,
ROUND(SUM(TXN),2) as usage_amount,
sum(AMOUNT) as AMOUNT,
COUNT(TXN) as nb_of_transactions,
ROUND(SUM(TXN)/COUNT(TXN),0) as amount_transaction_mean
FROM
(SELECT 
ID,
strftime('%H:%M:00',substr(APPLICATION_DATE, 1, 8)) as DATE,
STATUS,
MOTIVE,
CAST(INTEREST_RATE AS INTEGER) as INTEREST_RATE,
CAST(AMOUNT AS REAL) as AMOUNT,
CAST(CAT AS INTEGER) as CAT,
CAST(TXN AS NUMERIC) as TXN,
CAST(CP AS TEXT) as CP,
CAST(DELIVERY_SCORE AS TEXT) as DELIVERY_SCORE
FROM credit_card_data)
group by ID)
where
usage_amount>0
"""

usage = query(active)
usage
```

## Funnel Analysis

1.- Marketing campaigns reached 3341 potential users and there was only 2298 responses, i mean, 68% of total campaing.

2.- Risk Acceptance rate was 65%, and other words 35% of rejection rate. **RA = 1-(Rejected/risk)**

I think that there should be a more exhaustive analysis regarding which clients we are targeting to achieve a greater response and therefore have better conversion rates. Another important point is the percentage of physical cards delivered where it would be interesting to measure what type of segment, that is, **are they from a particular age group? Don't have other credit cards? etc.**

```python
waterfall = """
SELECT
COUNT(*) as count, 
STATUS
FROM
(SELECT 
ID,
strftime('%H:%M:00',substr(APPLICATION_DATE, 1, 8)) as DATE,
STATUS,
MOTIVE,
CAST(INTEREST_RATE AS INTEGER) as INTEREST_RATE,
CAST(AMOUNT AS REAL) as AMOUNT,
CAST(CAT AS INTEGER) as CAT,
CAST(TXN AS NUMERIC) as TXN,
CAST(CP AS TEXT) as CP,
CAST(DELIVERY_SCORE AS TEXT) as DELIVERY_SCORE
FROM credit_card_data)
where
STATUS is not null
group by STATUS
"""

waterfall = query(waterfall)

waterfall['order'] = np.where((waterfall['STATUS']=='RESPONSE') | (waterfall['STATUS']=='RISK'),1,
                             np.where(waterfall['STATUS']=='REJECTED',2,np.where(waterfall['STATUS']=='APPROVED',3,4)))

waterfall = waterfall.sort_values('order',ascending=True).set_index('order')


plt.bar(waterfall['STATUS'],waterfall['count'], color = 'orange')
xs = waterfall['STATUS'].astype(str)
ys = waterfall['count']

for x,y in zip(xs,ys):
    label = "{:.2f}".format(y)
    plt.annotate(label, 
                 (x,y), 
                 textcoords="offset points", 
                 xytext=(0,1), 
                 ha='center',
                fontsize=15) 
    
plt.title('Funnel Analysis - Waterfall',fontsize=20)
plt.rcParams["figure.figsize"] = (10,10)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

```python
reject_reason = """
SELECT
COUNT(*) as count, 
MOTIVE
FROM
(SELECT 
ID,
strftime('%H:%M:00',substr(APPLICATION_DATE, 1, 8)) as DATE,
STATUS,
MOTIVE,
CAST(INTEREST_RATE AS INTEGER) as INTEREST_RATE,
CAST(AMOUNT AS REAL) as AMOUNT,
CAST(CAT AS INTEGER) as CAT,
CAST(TXN AS NUMERIC) as TXN,
CAST(CP AS TEXT) as CP,
CAST(DELIVERY_SCORE AS TEXT) as DELIVERY_SCORE
FROM credit_card_data)
where
MOTIVE is not null
AND STATUS='REJECTED'
group by MOTIVE
"""

reason = query(reject_reason)
reason

plt.bar(reason['MOTIVE'],reason['count'], color = 'skyblue')
xm = reason['MOTIVE'].astype(str)
ym = reason['count']

for x,y in zip(xm,ym):
    label = "{:.2f}".format(y)
    plt.annotate(label, 
                 (x,y), 
                 textcoords="offset points", 
                 xytext=(0,1), 
                 ha='center',
                fontsize=15) 
    
plt.title('Reject Reason - Motive',fontsize=20)
plt.rcParams["figure.figsize"] = (10,10)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
nb_rejections = reason['count'].sum()
plt.text(.2, 200, f'Total Rejections: {nb_rejections}',
        fontsize = 13, color ="black")
plt.show()

```
